# Demo: Performance Evaluation with scikit-learn

## Package setup

```python
# Import base packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Setup plots
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 8
%config InlineBackend.figure_format = 'retina'
sns.set()
```

```python
# Import ML packages
from sklearn import metrics
```

## Accuracy

```python
# Define ground truth and prediction vectors
y_true = np.array([2, 0, 2, 2, 0, 1])
y_pred = np.array([0, 1, 2, 2, 0, 1])
```

```python
# Using scikit-learn's accuracy_score() function
acc = metrics.accuracy_score(y_true, y_pred)

print(f'Accuracy: {acc * 100:.2f}%')
```

## Precision, recall and f1-score

```python
# Using scikit-learn's precision_score() function
precision = metrics.precision_score(y_true, y_pred, average=None)

print(f'Precision: {precision}')
```

```python
# Using scikit-learn's recall_score() function
recall = metrics.recall_score(y_true, y_pred, average=None)

print(f'Recall: {recall}')
```

```python
print(metrics.classification_report(y_true, y_pred))
```

## Confusion matrix

```python
conf_mat = metrics.confusion_matrix(y_true, y_pred)

print(conf_mat)
```

```python
# Rather ugly for now, PR welcome :)
def plot_confusion_matrix(y_true, y_pred):
    df_cm = pd.DataFrame(metrics.confusion_matrix(y_true, y_pred))
    sns.heatmap(df_cm, annot=True, cmap='YlGnBu')
```
